# Geospatial Data Science Project

Salt Lake City, UT aims to achieve 80% carbon reduction under 2009 emission level. To achieve the goal, recently, the city adopted electrified transportation resolution to promote electric vehicle (EV) purchase, utilization, and infrastructures development. EV charging station, as the infrastructures of electrified transportation, plays an important role. This project will correspond this resolution, focusing on scalable analysis of charging stations magnitude and distribution. It will utilize CyberGIS to considering the impacts of traffic volumes, population. finally, it will generate two heatmaps for population and traffic volumes by using the cocept of MapReducer. <br><br>
The data sources in this project includes the city road network, traffic volume in 2040(projected by the department of transportation Utah), population in 2040. Finally, based on the concept of MapReducer, we can create two heatmaps of population and traffic volumes, showing the concentration of factors, which will give evidence to charging station site selection. 


```python
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
import shapefile
import pandas as pd

%matplotlib inline
import os
from geopandas import GeoSeries, GeoDataFrame, read_file, gpd
from pandas import Series
import IPython

# create basemap
fig=plt.figure(figsize=(20,10))

m = Basemap(llcrnrlon=-130,llcrnrlat=25,urcrnrlon=-64,urcrnrlat=50,
        projection='merc',lat_1=32,lat_2=45,lon_0=-95)
m.drawmapboundary(fill_color='White', zorder=-1)
m.drawcoastlines()
m.drawcountries(color='Black', linewidth=1)
m.drawstates(color = 'gray')
m.drawparallels(np.arange(10,90,20),labels=[1,1,0,1])
m.drawmeridians(np.arange(-180,180,30),labels=[1,1,0,1])
m.shadedrelief()

# scatter the city
m.scatter(-111.8910, 40.7608, s=500, color='Red', latlon=True)

# create title and legend
plt.title('Salt Lake City Location', fontsize = 25, ha='center')
print(plt.xlim(), plt.ylim())
```

```python
# load files vmt(average anual vehicle miles traveled) and city (salt lake city)
input_file = 'vmt'
input_file2 = 'city'
input_file3 = 'pop'
vmt=gpd.read_file(input_file)
city=gpd.read_file(input_file2)
pop=gpd.read_file(input_file3)
population=pop[['YEAR2040','geometry']]

# plot by layers
fig, ax = plt.subplots(figsize=(20,10))
ax.set_aspect('equal')
population.plot(ax=ax, column="YEAR2040",cmap='Greens', scheme='naturalbreaks', 
                        edgecolor='black',figsize=(20, 10), alpha=0.5, legend="true");
vmt.plot(ax=ax, column="F2040", cmap='YlOrRd', scheme='quantiles', 
           legend="true");
ax.set_title("Projected Traffic Volume in Salt Lake City (based on population projection 2040)", size=20)
```

# Geospatial Data Science Project – Part 3

```python
from IPython.core.display import HTML
import pandas as pd
from pandas import DataFrame
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import sys
import math
```

```python
pop = pd.read_csv('pop_point.csv')
vmt = pd.read_csv('vmt_point.csv')
```

### For population:

```python
# select column year2040, lon, lat.
df = pop.loc[:,['YEAR2040','lon','lat']]
df2 = vmt.loc[:,['F2040','lon','lat']]

# find min/max lan/lat.
max_lon = max(max(df['lon']),max(df2['lon']))
min_lon = min(min(df['lon']),min(df2['lon']))
max_lat = max(max(df['lat']),max(df2['lat']))
min_lat = min(min(df['lat']),min(df2['lat']))
print("max lon is:",max_lon)
print("min lon is:",min_lon)
print("max lat is:",max_lat)
print("min lat is:",min_lat)

# ser grid_size
grid_size = 0.005
print("grid size is:",grid_size)

# convert datafrane to list
list = df.values.tolist()
list2 = df2.values.tolist()
```

```python
def makeindex(record):
        lat=record[2]
        lon=record[1]
        row_index=int((lat-min_lat)/grid_size)
        col_index=int((lon-min_lon)/grid_size)
        value=record[0]
        
        return [(row_index,col_index),value]
```

```python
result = []
for row in list:
    values=makeindex(row)
    result.append(values)

result2 = []
for row in list2:
    values=makeindex(row)
    result2.append(values)
```

```python
result_merge=pd.DataFrame(data=result, columns=['index','value']).groupby('index').agg({"value":"sum"})
result_merge2=pd.DataFrame(data=result2, columns=['index','value']).groupby('index').agg({"value":"sum"})
```

```python
result_merge.to_csv('result_merge_pop.csv')
result_merge2.to_csv('result_merge_vmt.csv')
```

```python
pop_clean = pd.read_csv('result_merge_pop_clean.csv')
vmt_clean = pd.read_csv('result_merge_vmt_clean.csv')
```

```python
pop_matrix = pop_clean.pivot("row","col","value")
vmt_matrix = vmt_clean.pivot("row","col","value")
```

```python
sns.heatmap(pop_matrix, cmap='BuPu')
```

```python
sns.heatmap(vmt_matrix, cmap='BuPu')
```
